```python
fidap = fidap_client(api_key="e5bb3c8567d14dd28923d1cdbe0560a4")
```

```python
df = fidap.sql(
"""
select *
from fidap-301014.github.github_stars_pr_analysis
where REPO_NAME in ("facebook/react", "angular/angular", "vuejs/vue")
"""
)
```

```python
df
```

```python
import plotly.express as px
```

```python
stars = df.groupby(["REPO_NAME", "DATE"]).agg({'TOTAL_STARS': 'sum'}).reset_index()
stars.head()
```

```python
fig = px.line(stars, x="DATE", y="TOTAL_STARS", color="REPO_NAME")
fig.show()
```

```python
pull_requests = df.groupby(["REPO_NAME", "DATE"]).agg({'TOTAL_PULL_REQUESTS': 'sum'}).reset_index()
pull_requests.head()
```

```python
fig1 = px.line(pull_requests, x="DATE", y="TOTAL_PULL_REQUESTS", color="REPO_NAME")
fig1.show()
```
